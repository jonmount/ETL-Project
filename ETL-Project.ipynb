```python
import pandas as pd
from sqlalchemy import create_engine
import numpy as np
```

```python
#import file 1 
csv_file1 = "Resources/country_GDP.csv"
country_GDP = pd.read_csv(csv_file1)
country_GDP.head()




                                 
                                 
```

```python
#transpose year data 
country_GDP = country_GDP.melt(id_vars=["Country Name", "Country Code", "Indicator Name", "Indicator Code"], 
        var_name="Date")

country_GDP


country_GDP=country_GDP.set_index("Date")

country_GDP


#drop null GDP data 

country_GDP = country_GDP.dropna()

country_GDP


#rename Value column 

country_GDP = country_GDP.rename(columns={"value": "GDP"})

country_GDP
```

```python
csv_file2 = "Resources/country_population.csv"
country_pop = pd.read_csv(csv_file2)
country_pop.head()


#transpose year data 
country_pop = country_pop.melt(id_vars=["Country Name", "Country Code", "Indicator Name", "Indicator Code"], 
        var_name="Date")

country_pop


country_pop=country_pop.set_index("Date")

country_pop

#rename value 
country_pop = country_pop.rename(columns={"value": "Population"})

country_pop
```

```python
csv_file3 = "Resources/fertility_rate.csv"
fertility_rate = pd.read_csv(csv_file3)
fertility_rate.head()


fertility_rate = fertility_rate.melt(id_vars=["Country Name", "Country Code", "Indicator Name", "Indicator Code"], 
        var_name="Date")

fertility_rate


fertility_rate=fertility_rate.set_index("Date")

fertility_rate

#rename value 
fertility_rate = fertility_rate.rename(columns={"value": "Fertility Rate"})

fertility_rate

```

```python
csv_file4 = "Resources/life_expectancy.csv"
life_exp = pd.read_csv(csv_file4)
life_exp.head()


life_exp = life_exp.melt(id_vars=["Country Name", "Country Code", "Indicator Name", "Indicator Code"], 
        var_name="Date")

life_exp


life_exp=life_exp.set_index("Date")

life_exp


#rename value 
life_exp = life_exp.rename(columns={"value": "Life Expectancy"})

life_exp

```

```python
data_1 = pd.merge(life_exp, fertility_rate, on=('Date', 'Country Name'))



data_1


data_2 = pd.merge(country_pop, country_GDP, on=('Date', 'Country Name'))

data_2


all_data = pd.merge(data_1, data_2, on=('Date', 'Country Name'))

all_data
```

```python
country_data_final = all_data[["Country Name", "Life Expectancy", "Population", "Fertility Rate", "GDP"]]

#Add column GDP per Capita
country_data_final["GDP Per Capita"] = country_data_final["GDP"]/country_data_final["Population"]
country_data_final

#GDP in Billions
country_data_final["GDP in Billions"] = country_data_final["GDP"]/1000000000


#Round Columns
country_data_round = country_data_final.round({"Life Expectancy":2, "Fertility Rate":2, "GDP in Billions":2, "GDP Per Capita":2})
country_data_round


#Cleaned Dataframe
country_data_cleaned = country_data_round[["Country Name", "Life Expectancy", "Population", "Fertility Rate", "GDP in Billions", "GDP Per Capita"]]
country_data_cleaned



country_data_cleaned = country_data_cleaned.reset_index()

#Rename Table to fit postgres

Final_WB_DB = country_data_cleaned.rename(columns = {"Date":"year_wb","Country Name": "country_name", "Life Expectancy": "life_expectancy", "Population": "population", "Fertility Rate": "fertility_rate", "GDP in Billions": "gdp_billions", "GDP Per Capita": "gdp_per_capita"})
Final_WB_DB


Final_WB_DB



```

```python
#import world econ score data
SES = "Resources/GLOB-SES.csv"
world_econscore = pd.read_csv(SES, encoding='latin-1')


world_econscore.head()

world_econscore= world_econscore[["country", "year", "SES"]]
world_econscore.head()

world_econscore= world_econscore.rename(columns={"SES":"ses", "year": "year_ses"})

world_econscore.head()
                        
```
